```python
import pandas as pd
data = pd.read_csv('salary,satisfaction.csv',encoding='latin1')
```

```python
null_columns = data.columns[data.isnull().any()]
sum = data[null_columns].isnull().sum()
print(sum.sort_values(ascending=False))
print(sum.shape)
```

```python
data.head(5)
```

```python
#checking and removing duplicate entries
print(data.shape)
data.drop_duplicates(subset='id')
print(data.shape)
```

```python
lst = []

#converting all values into float
for i  in data['LOAN_AMT'].astype(str):
  i = str(i).replace(',','')
  lst.append(float(i[1:-1]))

#changing inplace of original data
data['LOAN_AMT'] = lst
```

```python
#im using label encoder for encoding rather than one hot encoding 
#because it create more columns and require more computational power

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

columns_needed = ['FormalEducation','UndergradMajor','CompanySize','YearsCoding',
'YearsCodingProf','HopeFiveYears','JobSearchStatus',
'LastNewJob','UpdateCV','TimeFullyProductive','TimeAfterBootcamp',
'HackathonReasons','AgreeDisagree1','AgreeDisagree2','AgreeDisagree3',
'OperatingSystem','CheckInCode','AIDangerous','AIInteresting','AIResponsible',
'AIFuture','Age','term','loan_status','purpose','title','emp_length']


for i in columns_needed:
  data[i] = le.fit_transform(data[i].astype(str))
```

```python
target = data['Target_Satisfied']
features = data[columns_needed]
```

```python
# features.fillna(0)
len(columns_needed)
```

```python
#classification
#importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score

#splitting data into training and testing set
X_train, X_test, y_train, y_test = train_test_split(features,target,test_size = 0.2,random_state = 7)

#scaling data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)


#model building
class_model = RandomForestClassifier(n_estimators = 500,criterion= 'gini')
class_model.fit(X_train,y_train)
class_pred = class_model.predict(X_test)

#printing model scores
print('classification model score for training set is ==>',class_model.score(X_train,y_train))
print('classification model score for testing set is  ==>',class_model.score(X_test,y_test),'\n','\n')

#printing root mean squared error
print('f1 score for classification model predictions is ==>',f1_score(y_test,class_pred))
```
