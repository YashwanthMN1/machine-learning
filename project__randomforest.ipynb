```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('salary,satisfaction.csv',encoding='latin1')
```

```python
#checking and removing duplicate entries
print(data.shape)
data.drop_duplicates(subset='id')
print(data.shape)
```

```python
data['LOAN_AMT'] = data['LOAN_AMT'].replace({'$': '', ',': ''}, regex=True).astype(float)
```

```python
null_columns = data.columns[data.isnull().any()]
sum = data[null_columns].isnull().sum()
print(sum)
print(sum.shape)
```

```python
data.fillna(0)
```

```python
data1 = data
```

```python
def delimeter_counter(col):
  temp_lst = []
  for i in col:
    if i == 0:
      temp_lst.append(0)
      pass
    elif str(i).count(';') == 0 :
      temp_lst.append(1)
    elif str(i).count(';') > 0 :
      c = str(i).count(';')
      temp_lst.append(int(c+1))
  return temp_lst

columns_of_delimeter = ['LanguageWorkedWith','LanguageDesireNextYear','DatabaseWorkedWith',
'DatabaseDesireNextYear',
'PlatformWorkedWith',
'PlatformDesireNextYear',
'FrameworkWorkedWith',
'FrameworkDesireNextYear',
'IDE',
'Methodology']


for x in columns_of_delimeter:
  l = delimeter_counter(data1[x].values)
  data1[x] = l


#data1
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
le = LabelEncoder()
ohe = OneHotEncoder()
```

```python
columns_for_le = ['FormalEducation','UndergradMajor','CompanySize','YearsCoding',
'YearsCodingProf','HopeFiveYears','JobSearchStatus',
'LastNewJob','UpdateCV','TimeFullyProductive','TimeAfterBootcamp',
'HackathonReasons','AgreeDisagree1','AgreeDisagree2','AgreeDisagree3',
'OperatingSystem','CheckInCode','AIDangerous','AIInteresting','AIResponsible',
'AIFuture','Age','term','loan_status','purpose','title','emp_length']

# columns_for_ohe = []'''

for i in columns_for_le:
  data1[i] = le.fit_transform(data1[i].astype(str))
  print(i)
```

```python
# wanted_columns = []
# features  = data[wanted_columns]
target_classification = data['Target_Satisfied']
target_regression = data['Target_Salary']
```

```python
co = data1.corr()
print(co['Target_Satisfied'].sort_values())
print(co['Target_Salary'].sort_values())
```

```python
l = columns_of_delimeter + columns_for_le

features = data1[l]
target_class = data1['Target_Satisfied']
target_reg = data1['Target_Salary']
```

```python
from sklearn.model_selection import train_test_split
xtrain_class,xtest_class,ytrain_class,ytest_class = train_test_split(features,target_class,test_size = 0.2,random_state = 7)
xtrain_reg,xtest_reg,ytrain_reg,ytest_reg = train_test_split(features,target_reg,test_size = 0.2,random_state = 7)
```

```python
#classification
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
class_model = RandomForestClassifier()
class_model.fit(xtrain_class,ytrain_class)
class_pred = class_model.predict(xtest_class)
print(class_model.score(xtrain_class,ytrain_class))
print(class_model.score(xtest_class,ytest_class))
```

```python
from sklearn.metrics import f1_score
print(f1_score(ytest_class,class_pred))
```

```python
#regression
from sklearn.ensemble import RandomForestRegressor
reg_model = RandomForestRegressor()
reg_model.fit(xtrain_reg,ytrain_reg)
reg_predictions = reg_model.predict(xtest_reg)
print(reg_model.score(xtrain_reg,ytrain_reg))
print(reg_model.score(xtest_reg,ytest_reg))
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(ytest_reg,reg_predictions, squared=False))
```
